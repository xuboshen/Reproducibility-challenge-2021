```python
import matplotlib.pyplot as plt
import numpy as np
acc_last_step = np.array([79.52, 82.04, 81.83, 75.48, 83.94])
acc_best_val = np.array([80.79, 81.51, 83.04, 75.45, 79.27])
alpha = ['res1', 'res12', 'res123', 'res1234', 'res2']
x = np.arange(len(alpha))
print(x)
width = 0.25
fig, ax = plt.subplots(figsize = (10, 8))
ax.set_xticks(x)
ax.set_ylim(75, 85)
ax.bar(x - width/2, acc_last_step, width, label = 'Last Step')
ax.bar(x + width/2, acc_best_val, width, label = 'Best Val')
ax.plot(x - width/2, acc_last_step, color='black', marker = 'o', markeredgecolor='black')
ax.plot(x + width/2, acc_best_val, color='black', marker = 'o', markeredgecolor='black')
ax.set_xticklabels(alpha)
ax.legend()
ax.set_xlabel(r'The Places to Apply MixStyle')
ax.set_ylabel(r'Accuracy(/%)')
ax.set_title('PACS')

ax.grid()
fig.savefig('accuracy_with_res.png', dpi=200)
```
